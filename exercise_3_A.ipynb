**Instructions:**

Build a Simple Linear Regression model to study the linear relationship between the years of experience and salary.


1. Import the necessary packages

```python
# put your answer here

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

2. Load the salary.csv file

```python
# put your answer here

df = pd.read_csv("salary.csv")
```

3. Print the first 5 rows of the dataset

```python
# put your answer here

df.head(5)
```

4. Display the general information of the dataframe

```python
# put your answer here

df.info()
```

5. Visualize the data by creating a scatter plot

```python
# put your answer here
df.plot(x="YearsExperience", y="Salary", kind="scatter")
```

6. Split the dataset into a training set and test set

```python
# put answer here

X = df["YearsExperience"].values
y = df["Salary"].values

```

7. Create an new instance of a linear regression model

8. Train and test the model using the training set and test set

```python
# put your answer here

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size= 0.33, random_state=42)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# put your answer here
mylinearmodel = LinearRegression()

mylinearmodel.fit(X_train, y_train)

y_pred = mylinearmodel.predict(X_test)
```

9. Print the equation of the line

- Display the slope
- Display the intercept
- Display the equation of the line

```python
# put your answer here

#for slope
m = mylinearmodel.coef_[0]
m = m[0]

#for intercept
b = mylinearmodel.intercept_[0]

print ("m = ", m)
print ("b = ", b)
print ("y = ", m, " * + ", b)
```

10. Draw the Regression Line

```python
# put your answer here

plt.scatter (X, y, color = 'pink', label ='Data Points')
plt.plot (X_test, y_pred, color = 'yellow', linewidth = 3, label = 'Regression Line')
plt.title ('Relationship between the Years of Experience & Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(loc=4)
plt.show()
```

11. Compute the Sum of Square Error (SSE)

```python
# put your answer here
y_test = y_test.reshape(-1)

y_pred = y_pred.reshape(-1)

data = pd.DataFrame({'Actual Salary': y_test, 'Predicted Salary': y_pred})

data.head()
```

```python
sose = str(np.sum(np.square(data['Predicted Salary'] - data['Actual Salary'])))

print ('Residual sum of square is: ', sose)
```
